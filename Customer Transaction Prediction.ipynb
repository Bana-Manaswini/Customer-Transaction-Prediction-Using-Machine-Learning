# Task 2:

### Feature Selection

```python
X = data.drop(['ID_code','target'], axis=1)
y = data['target']                 # drop the ID_code and target column so that we dont want that columns to train the modal
```

```python
y                                  # taking the target data a target variable
```

### Model Creation

```python
## Step 1 Creating training and testing data.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train , y_test = train_test_split(X,y,test_size=0.30,random_state=2)
```

```python
y_test
```

```python
## Step 2 Model creation
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(X_train,y_train)  ## training
```

```python
## Step 3 Prediction
y_pred=clf.predict(X_test)            # predicting the y_predict by using x_test
```

```python
y_pred                                
```

```python
y_pred_prob=clf.predict_proba(X_test)   # probability of y_predict by using x_test
```

```python
y_pred_prob
```

```python
data.target.value_counts()              # no. of transactions for the present data
```

### Model Evalution

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)                                 # to evaluate the performance of a classification model
```

```python
recall=recall_score(y_test,y_pred)
recall                                    # metric that measures how well a model identifies positive cases
```

```python
precision=precision_score(y_test,y_pred)
precision                                 # the quality of a positive prediction made by the model
```

```python
f1score=f1_score(y_test,y_pred)
f1score                                   # that measures a model's accuracy by combining its precision and recall scores
```

```python
cr=classification_report(y_test,y_pred)
print(cr)                                  # a text summary that shows the main metrics for each class of a machine learning model
```
